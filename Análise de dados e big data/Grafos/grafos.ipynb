```python
import networkx as nx
from networkx.classes import nodes

G = nx.Graph()
```

```python
# Adicionando Vertices
G.add_node(1)
G.add_node(2)
```

```python
# Visualizando os vertices
print(G.nodes())
```

```python
# Adicionando Arestas
G.add_edge(1, 2)
# Total de vertices e arestas
n_vertices = G.number_of_nodes()
n_arestas = G.number_of_edges()
print(f"Total de vertices: {n_vertices}\nTotal de arestas: {n_arestas}")

G.add_edge(4, 5)
print(G.nodes())
print(G.edges())
```

```python
# Desenhando
nx.draw(G, with_labels=True, node_size=1200,
        node_color="red")
```

```python
# Mais vertices;
G.add_node(1)
G.add_node(2)
print(G.nodes())
```

```python
# Remove o nó 2
G.remove_node(2)
print(G.nodes())
```

```python
# Adicionando arestas sem necessidade de adicionar os nós
G.add_edge(3, 4)
G.add_edge(1, 3)
G.add_edge(5, 6)
G.add_edge(5, 9)
G.add_edge(7, 8)
G.add_edge(1, 8)
G.add_edge(1, 4)
nx.draw(G, with_labels=True, node_size=1200,
        node_color="red", edge_color="blue", width=2, alpha=0.5, font_size=16)
```

```python
# Subgrafos, são as partes do grafo que não são conectadas entre
subgrafos =[G.subgraph(s) for s in
            nx.connected_components(G)]

print(len(subgrafos))
```

```python
nx.draw(subgrafos[0], with_labels=True, node_size=800,)
```

```python
nx.draw(subgrafos[1], with_labels=True, node_size=800,
        node_color='green')
```
