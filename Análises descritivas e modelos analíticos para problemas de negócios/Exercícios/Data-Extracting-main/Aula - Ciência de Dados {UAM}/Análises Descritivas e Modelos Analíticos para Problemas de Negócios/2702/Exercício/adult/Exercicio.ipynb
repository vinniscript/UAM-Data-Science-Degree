```python
import pandas as pd
```

```python
df_lenses  = pd.read_csv('adult.csv', header=None, names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country'])
df_lenses
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df_lenses['age'] = encoder.fit_transform(df_lenses['age'])
df_lenses['workclass'] = encoder.fit_transform(df_lenses['workclass'])
df_lenses['fnlwgt'] = encoder.fit_transform(df_lenses['fnlwgt'])
df_lenses['education'] = encoder.fit_transform(df_lenses['education'])
df_lenses['education-num'] = encoder.fit_transform(df_lenses['education-num'])
df_lenses['marital-status'] = encoder.fit_transform(df_lenses['marital-status'])
df_lenses['occupation'] = encoder.fit_transform(df_lenses['occupation'])
df_lenses['relationship'] = encoder.fit_transform(df_lenses['relationship'])
df_lenses['race'] = encoder.fit_transform(df_lenses['race'])
df_lenses['hours-per-week'] = encoder.fit_transform(df_lenses['hours-per-week'])
df_lenses.head()
```
